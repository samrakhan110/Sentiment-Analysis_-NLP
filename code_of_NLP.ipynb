Importing modules

```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('stopwords')

nltk.download('wordnet')

nltk.download('punkt')
```

Reading the file

```python
# Load the text files
with open('speech_of_ 2021.txt', 'r') as speech:
    text1 = speech.read()


with open('speech_of_2022.txt', 'r') as speech:
    text2 = speech.read()
```

# New section

tokenization

```python
# Tokenize the text

tokens1= word_tokenize(text1)
tokens2= word_tokenize(text2)

print(tokens1)
print(tokens2)
```

lower the words

```python
# Converting into lowercase
tokens1 = [token.lower() for token in tokens1]
tokens2 = [token.lower() for token in tokens2]
print(tokens1)
print(tokens2)
```

stop words

```python
# Remove stopwords
stopwords =set(stopwords.words('english'))  #unimportant words
tokens1 = [word for word in tokens1 if word not in stopwords]
tokens2 = [word for word in tokens2 if word not in stopwords]
print(stopwords)
```

lemmatization

```python
# Lemmatization
lemmatizer = nltk.WordNetLemmatizer()
tokens1 = [lemmatizer.lemmatize(word) for word in tokens1]
tokens2 = [lemmatizer.lemmatize(word) for word in tokens2]
print(tokens1)
print(tokens2)
```

common words

```python
# Get the common words
common_words = set(tokens1) & set(tokens2)
print(common_words)
```

counting the frequency of words

```python
# Counting the frequency of words
word_freq1 = nltk.FreqDist(tokens1)
word_freq2 = nltk.FreqDist(tokens2)
# Printing common words and number of times the word has occured
for word in common_words:
    freq1 = word_freq1[word]
    freq2 = word_freq2[word]
    print(f"Word: {word}, Frequency in File 1: {freq1}, Frequency in File 2: {freq2}")
```

word cloud

```python
speech1 = WordCloud(width=1000, height=600).generate(' '.join(tokens1))
speech2 = WordCloud(width=1000, height=600).generate(' '.join(tokens2))
```

displaying word cloud

```python
plt.figure(figsize=(16, 8))    #for first speech
plt.subplot(1, 2, 1)
plt.imshow(speech1, interpolation='bilinear')
plt.title('Word Cloud for speech 1')
plt.axis("off")

plt.subplot(1, 2, 2)              # for second speech
plt.imshow(speech2, interpolation='bilinear')
plt.title('Word Cloud for speech 2')
plt.axis("off")
plt.show()
```

Common words from both speeches


```python
# Creating a word cloud of common words between speech 1 and speech 2
plt.figure(figsize=(10, 4))
wordcloud = WordCloud().generate(' '.join(common_words))

# Showing a word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Sentiment Analysis


```python
#sentimental analysis
Senti_ment = SentimentIntensityAnalyzer()
# for speech 1
sentiment1 = Senti_ment.polarity_scores(text1)
print(sentiment1)
# for speech 2
sentiment2 = Senti_ment.polarity_scores(text2)
print(sentiment2)
```

```python
# Extracting negative and positive words
negative_words = [word for word in tokens1 + tokens2 if Senti_ment.polarity_scores(word)['compound'] < 0]
positive_words = [word for word in tokens1 + tokens2 if Senti_ment.polarity_scores(word)['compound'] > 0]

min_length = min(len(negative_words), len(positive_words))
negative_words = negative_words[:min_length]
positive_words = positive_words[:min_length]
#creating data frame for sentiment analysis
df = pd.DataFrame({'Negative Words': negative_words, 'Positive Words': positive_words})
print(df)
```

```python
wordcloud = WordCloud().generate(' '.join(negative_words))

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wordcloud = WordCloud().generate(' '.join(positive_words))

# showing the word cloud of positive word
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
